```python
import numpy as np,pandas as pd
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

exam_data=pd.DataFrame(exam_data)
exam_data=exam_data.set_index([labels])
print(exam_data.head(4),'\n')
exam_data=exam_data.dropna()
print(exam_data.iloc[:,0:2],'\n')
k= {'name': ['Suresh'],
'score': [15.5],
'attempts': [1],
'qualify': ['yes']}
k=pd.DataFrame(k)
exam_data=pd.concat([exam_data,k])
print(exam_data,'\n')
exam_data.drop("attempts",inplace=True,axis=1)
print(exam_data,'\n')
Score=exam_data['score']
Success=list()
for i in Score:
    if i>=10:
        Success.append(1)
    else:
        Success.append(0)
exam_data['Success']=Success
print(exam_data)
exam_data.to_csv('my_data.csv')
exam_data=pd.read_csv('my_data.csv')
```
